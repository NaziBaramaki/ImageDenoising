```python
import numpy as np
from skimage.filters import gaussian
from skimage.io import imread, imshow, show
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Load the initial "noisy office" image
initial_image = imread('office_noisy.png')
```

```python
# Solve the diffusion PDE for different values of the diffusion coefficient
for d in [1, 5, 10]:
    output_image = gaussian(initial_image, sigma=np.sqrt(10 * d))
    print(f"Output image with diffusion coefficient d = {d}:")
    imshow(output_image)
    show()
```
